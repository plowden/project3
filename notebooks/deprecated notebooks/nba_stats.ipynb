```python
from basketball_reference_web_scraper import client
import json
import pandas as pd
```

```python
competition_year = []
player_id = []
name = []
position = []
age = []
team = []
games_played = []
games_started = []
minutes_played = []
points = []
made_field_goals = []
attempted_field_goals = []
made_three_pointers = []
attempted_three_pointers = []
made_free_throws = []
attempted_free_throws = []
offensive_rebounds = []
defensive_rebounds = []
assists = []
steals = []
blocks = []
turnovers = []
personal_fouls = []

for year in range(2000, 2020):
    stats = client.players_season_totals(season_end_year=year)
    for result in stats:
        competition_year.append(year)
        player_id.append(result['slug'])
        name.append(result['name'])
        position.append(result['positions'][0])
        age.append(result['age'])
        team.append(result['team'])
        games_played.append(result['games_played'])
        games_started.append(result['games_started'])
        minutes_played.append(result['minutes_played'])
        points.append(result['points'])
        made_field_goals.append(result['made_field_goals'])
        attempted_field_goals.append(result['attempted_field_goals'])
        made_three_pointers.append(result['made_three_point_field_goals'])
        attempted_three_pointers.append(result['attempted_three_point_field_goals'])
        made_free_throws.append(result['made_free_throws'])
        attempted_free_throws.append(result['attempted_free_throws'])
        offensive_rebounds.append(result['offensive_rebounds'])
        defensive_rebounds.append(result['defensive_rebounds'])
        assists.append(result['assists'])
        steals.append(result['steals'])
        blocks.append(result['blocks'])
        turnovers.append(result['turnovers'])
        personal_fouls.append(result['personal_fouls'])
```

```python
# set up for loop to gather nba advanced data from years 2000-2019
player_id_2 = []
player_efficiency_rating = []
true_shooting_percentage = []
three_point_attempt_rate = []
free_throw_attempt_rate = []
offensive_rebound_percentage = []
defensive_rebound_percentage = []
total_rebound_percentage = []
assist_percentage = []
steal_percentage = []
block_percentage = []
turnover_percentage = []
usage_percentage = []
offensive_win_shares = []
defensive_win_shares = []
win_shares = []
win_shares_per_48_minutes = []
offensive_box_plus_minus = []
defensive_box_plus_minus = []
box_plus_minus = []
value_over_replacement_player = []

for year in range(2000, 2020):
    stats = client.players_advanced_season_totals(season_end_year=year)
    for result in stats:
        player_id_2.append(result['slug'])
        player_efficiency_rating.append(result['player_efficiency_rating'])
        true_shooting_percentage.append(result['true_shooting_percentage'])
        three_point_attempt_rate.append(result['three_point_attempt_rate'])
        free_throw_attempt_rate.append(result['free_throw_attempt_rate'])
        offensive_rebound_percentage.append(result['offensive_rebound_percentage'])
        defensive_rebound_percentage.append(result['defensive_rebound_percentage'])
        total_rebound_percentage.append(result['total_rebound_percentage'])
        assist_percentage.append(result['assist_percentage'])
        steal_percentage.append(result['steal_percentage'])
        block_percentage.append(result['block_percentage'])
        turnover_percentage.append(result['turnover_percentage'])
        usage_percentage.append(result['usage_percentage'])
        offensive_win_shares.append(result['offensive_win_shares'])
        defensive_win_shares.append(result['defensive_win_shares'])
        win_shares.append(result['win_shares'])
        win_shares_per_48_minutes.append(result['win_shares_per_48_minutes'])
        offensive_box_plus_minus.append(result['offensive_box_plus_minus'])
        defensive_box_plus_minus.append(result['defensive_box_plus_minus']) 
        box_plus_minus.append(result['box_plus_minus'])
        value_over_replacement_player.append(result['value_over_replacement_player'])
```

```python
columns = ['year', 'player_id', 'name', 'position', 'age', 'team', 'games_played', 'games_started',
          'minutes_played', 'points', 'made_field_goals', 'attempted_field_goals', 'made_three_pointers',
          'attempted_three_pointers', 'made_free_throws', 'attempted_free_throws', 'offensive_rebounds',
          'defensive_rebounds', 'assists', 'steals', 'blocks', 'turnovers', 'personal_fouls', 'player_id_2', 'player_efficiency_rating', 'true_shooting_percentage', 'three_point_attempt_rate',
           'free_throw_attempt_rate', 'offensive_rebound_percentage', 'defensive_rebound_percentage', 
           'total_rebound_percentage', 'assist_percentage','steal_percentage', 'block_percentage', 
           'turnover_percentage', 'usage_percentage', 'offensive_win_shares', 'defensive_win_shares', 'win_shares',
           'win_shares_per_48_minutes', 'offensive_box_plus_minus', 'defensive_box_plus_minus', 'box_plus_minus',
           'value_over_replacement_player']

complete_nba_stats = pd.DataFrame(columns=columns)
```

```python
complete_nba_stats['year'] = competition_year
complete_nba_stats['player_id'] = player_id
complete_nba_stats['name'] = name
complete_nba_stats['position'] = position
complete_nba_stats['age'] = age
complete_nba_stats['team'] = team
complete_nba_stats['games_played'] = games_played
complete_nba_stats['games_started'] = games_started
complete_nba_stats['minutes_played'] = minutes_played
complete_nba_stats['points'] = points
complete_nba_stats['made_field_goals'] = made_field_goals
complete_nba_stats['attempted_field_goals'] = attempted_field_goals
complete_nba_stats['made_three_pointers'] = made_three_pointers
complete_nba_stats['attempted_three_pointers'] = attempted_three_pointers
complete_nba_stats['made_free_throws'] = made_free_throws
complete_nba_stats['attempted_free_throws'] = attempted_free_throws
complete_nba_stats['offensive_rebounds'] = offensive_rebounds
complete_nba_stats['defensive_rebounds'] = defensive_rebounds
complete_nba_stats['assists'] = assists
complete_nba_stats['steals'] = steals
complete_nba_stats['blocks'] = blocks
complete_nba_stats['turnovers'] = turnovers
complete_nba_stats['personal_fouls'] = personal_fouls
complete_nba_stats['player_id_2'] = player_id_2
complete_nba_stats['player_efficiency_rating'] = player_efficiency_rating
complete_nba_stats['true_shooting_percentage'] = true_shooting_percentage
complete_nba_stats['three_point_attempt_rate'] = three_point_attempt_rate
complete_nba_stats['free_throw_attempt_rate'] = free_throw_attempt_rate
complete_nba_stats['offensive_rebound_percentage'] = offensive_rebound_percentage
complete_nba_stats['defensive_rebound_percentage'] = defensive_rebound_percentage
complete_nba_stats['total_rebound_percentage'] = total_rebound_percentage
complete_nba_stats['assist_percentage'] = assist_percentage
complete_nba_stats['steal_percentage'] = steal_percentage
complete_nba_stats['block_percentage'] = block_percentage
complete_nba_stats['turnover_percentage'] = turnover_percentage
complete_nba_stats['usage_percentage'] = usage_percentage
complete_nba_stats['offensive_win_shares'] = offensive_win_shares
complete_nba_stats['defensive_win_shares'] = defensive_win_shares
complete_nba_stats['win_shares'] = win_shares
complete_nba_stats['win_shares_per_48_minutes'] = win_shares_per_48_minutes
complete_nba_stats['offensive_box_plus_minus'] = offensive_box_plus_minus
complete_nba_stats['defensive_box_plus_minus'] = defensive_box_plus_minus
complete_nba_stats['box_plus_minus'] = box_plus_minus
complete_nba_stats['value_over_replacement_player'] = value_over_replacement_player
```
