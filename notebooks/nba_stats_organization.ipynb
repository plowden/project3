```python
import pandas as pd 
```

```python
nba_stats = pd.read_csv("../data/nba_stats.csv")
```

```python
del nba_stats["age"]
```

```python
# Groupby year and player 
nba_grouped = nba_stats.groupby(["year", "name", "player_id"])
```

```python
nba_sum_grouped = nba_grouped.sum()
```

```python
# Calculations

# average minutes
nba_sum_grouped["average_minutes"] = nba_sum_grouped["minutes_played"] / nba_sum_grouped["games_played"]
# average points 
nba_sum_grouped["average_points"] = nba_sum_grouped["points"] / nba_sum_grouped["games_played"]
# field goal percentage
nba_sum_grouped["field_goal_percentage"] = nba_sum_grouped["made_field_goals"] / nba_sum_grouped["attempted_field_goals"]
# three point percentage
nba_sum_grouped["three_point_percentage"] = nba_sum_grouped["made_three_pointers"] / nba_sum_grouped["attempted_three_pointers"]
# free throw percentage 
nba_sum_grouped["free_throw_percentage"] = nba_sum_grouped["made_free_throws"] / nba_sum_grouped["attempted_free_throws"]
# average rebounds 
nba_sum_grouped["average_rebounds"] = (nba_sum_grouped["offensive_rebounds"] + nba_sum_grouped["defensive_rebounds"]) / nba_sum_grouped["games_played"]
# average offenive rebounds
nba_sum_grouped["average_offensive_rebounds"] = nba_sum_grouped["offensive_rebounds"] / nba_sum_grouped["games_played"]
# average defensive rebounds 
nba_sum_grouped["average_defensive_rebounds"] = nba_sum_grouped["defensive_rebounds"] / nba_sum_grouped["games_played"]
# average assists
nba_sum_grouped["average_assists"] = nba_sum_grouped["assists"] / nba_sum_grouped["games_played"]
# average steals
nba_sum_grouped["average_steals"] = nba_sum_grouped["steals"] / nba_sum_grouped["games_played"]
# average blocks
nba_sum_grouped["average_blocks"] = nba_sum_grouped["blocks"] / nba_sum_grouped["games_played"]
# average turnovers 
nba_sum_grouped["average_turnovers"] = nba_sum_grouped["turnovers"] / nba_sum_grouped["games_played"]
# average personal fouls 
nba_sum_grouped["average_personal_fouls"] = nba_sum_grouped["personal_fouls"] / nba_sum_grouped["games_played"]

nba_sum_grouped.head()
```

```python
nba_sum_grouped = nba_sum_grouped.reset_index()
nba_sum_grouped
```

```python
# Import id_map_asg.csv
id_map_asg = pd.read_csv("../data/id_map_asg.csv")
```

```python
# Create list of all-stars
all_stars = []

for row in id_map_asg.itertuples():
    if row[5] == 1.0:
        all_stars.append(row[3])
```

```python
# build helper functions

def game_check(variable):
    result = []

    for x in variable:
        if x > 40:
            result.append(1)
        else:
            result.append(0)

    if sum(result) > 4:
        return True
    else:
        return False

def starter_minutes(variable):
    result = []
    
    for x in variable:
        if x >= 24:
            result.append(1)
        else: 
            result.append(0)
    
    if sum(result) > 4:
        return True
    else:
        return False
    
def role_minutes(variable):
    result = []
    
    for x in variable:
        if x < 24 and x >= 12:
            result.append(1)
        else: 
            result.append(0)
    
    if sum(result) > 2:
        return True
    else:
        return False
```

```python
# Set up for loop to create player classification
player_id = nba_sum_grouped["player_id"]
classification = []

for player in player_id:
    player_stats = nba_sum_grouped.loc[nba_sum_grouped["player_id"] == player, :]
    games_played = player_stats["games_played"]
    games = game_check(games_played)
    minutes = player_stats["average_minutes"]
    minutes_starter = starter_minutes(minutes)
    minutes_role = role_minutes(minutes)
    
    if player in all_stars:
        classification.append("All-Star") 
    elif len(player_stats) > 4 and games == True and minutes_starter == True:
        classification.append("Starter")
    elif len(player_stats) > 2 and games == True and minutes_role == True:
        classification.append("Role Player")
    else:
        classification.append("Bust")
```

```python
# Add Classification Column
nba_sum_grouped["classification"] = classification
```

```python
print(nba_sum_grouped["classification"].value_counts())
```

```python
nba_sum_grouped.to_csv("../data/nba_data_classified.csv", index=False)
```
