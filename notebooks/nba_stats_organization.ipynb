```python
import pandas as pd 
```

```python
nba_stats = pd.read_csv("../data/nba_stats.csv")
```

```python
del nba_stats["team"]
del nba_stats["age"]
```

```python
# Groupby year and player 
nba_grouped = nba_stats.groupby(["year", "name"])
```

```python
nba_sum_grouped = nba_grouped.sum()
nba_sum_grouped.head(50)
```

```python
# Calculations

# average minutes
nba_sum_grouped["average_minutes"] = nba_sum_grouped["minutes_played"] / nba_sum_grouped["games_played"]
# average points 
nba_sum_grouped["average_points"] = nba_sum_grouped["points"] / nba_sum_grouped["games_played"]
# field goal percentage
nba_sum_grouped["field_goal_percentage"] = nba_sum_grouped["made_field_goals"] / nba_sum_grouped["attempted_field_goals"]
# three point percentage
nba_sum_grouped["three_point_percentage"] = nba_sum_grouped["made_three_pointers"] / nba_sum_grouped["attempted_three_pointers"]
# free throw percentage 
nba_sum_grouped["free_throw_percentage"] = nba_sum_grouped["made_free_throws"] / nba_sum_grouped["attempted_free_throws"]
# average rebounds 
nba_sum_grouped["average_rebounds"] = (nba_sum_grouped["offensive_rebounds"] + nba_sum_grouped["defensive_rebounds"]) / nba_sum_grouped["games_played"]
# average offenive rebounds
nba_sum_grouped["average_offensive_rebounds"] = nba_sum_grouped["offensive_rebounds"] / nba_sum_grouped["games_played"]
# average defensive rebounds 
nba_sum_grouped["average_defensive_rebounds"] = nba_sum_grouped["defensive_rebounds"] / nba_sum_grouped["games_played"]
# average assists
nba_sum_grouped["average_assists"] = nba_sum_grouped["assists"] / nba_sum_grouped["games_played"]
# average steals
nba_sum_grouped["average_steals"] = nba_sum_grouped["steals"] / nba_sum_grouped["games_played"]
# average blocks
nba_sum_grouped["average_blocks"] = nba_sum_grouped["blocks"] / nba_sum_grouped["games_played"]
# average turnovers 
nba_sum_grouped["average_turnovers"] = nba_sum_grouped["turnovers"] / nba_sum_grouped["games_played"]
# average personal fouls 
nba_sum_grouped["average_personal_fouls"] = nba_sum_grouped["personal_fouls"] / nba_sum_grouped["games_played"]

nba_sum_grouped.head()
```

```python
nba_sum_grouped.to_csv("../data/nba_sum_grouped.csv", index = False)
```
