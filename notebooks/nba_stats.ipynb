```python
from basketball_reference_web_scraper import client
import json
import pandas as pd
```

```python
competition_year = []
player_id = []
name = []
position = []
age = []
team = []
games_played = []
games_started = []
minutes_played = []
points = []
made_field_goals = []
attempted_field_goals = []
made_three_pointers = []
attempted_three_pointers = []
made_free_throws = []
attempted_free_throws = []
offensive_rebounds = []
defensive_rebounds = []
assists = []
steals = []
blocks = []
turnovers = []
personal_fouls = []

for year in range(2000, 2020):
    stats = client.players_season_totals(season_end_year=year)
    for result in stats:
        competition_year.append(year)
        player_id.append(result['slug'])
        name.append(result['name'])
        position.append(result['positions'][0])
        age.append(result['age'])
        team.append(result['team'])
        games_played.append(result['games_played'])
        games_started.append(result['games_started'])
        minutes_played.append(result['minutes_played'])
        points.append(result['points'])
        made_field_goals.append(result['made_field_goals'])
        attempted_field_goals.append(result['attempted_field_goals'])
        made_three_pointers.append(result['made_three_point_field_goals'])
        attempted_three_pointers.append(result['attempted_three_point_field_goals'])
        made_free_throws.append(result['made_free_throws'])
        attempted_free_throws.append(result['attempted_free_throws'])
        offensive_rebounds.append(result['offensive_rebounds'])
        defensive_rebounds.append(result['defensive_rebounds'])
        assists.append(result['assists'])
        steals.append(result['steals'])
        blocks.append(result['blocks'])
        turnovers.append(result['turnovers'])
        personal_fouls.append(result['personal_fouls'])
```

```python
columns = ['year', 'player_id', 'name', 'position', 'age', 'team', 'games_played', 'games_started',
          'minutes_played', 'points', 'made_field_goals', 'attempted_field_goals', 'made_three_pointers',
          'attempted_three_pointers', 'made_free_throws', 'attempted_free_throws', 'offensive_rebounds',
          'defensive_rebounds', 'assists', 'steals', 'blocks', 'turnovers', 'personal_fouls']

nba_stats = pd.DataFrame(columns=columns)
```

```python
nba_stats['year'] = competition_year
nba_stats['player_id'] = player_id
nba_stats['name'] = name
nba_stats['position'] = position
nba_stats['age'] = age
nba_stats['team'] = team
nba_stats['games_played'] = games_played
nba_stats['games_started'] = games_started
nba_stats['minutes_played'] = minutes_played
nba_stats['points'] = points
nba_stats['made_field_goals'] = made_field_goals
nba_stats['attempted_field_goals'] = attempted_field_goals
nba_stats['made_three_pointers'] = made_three_pointers
nba_stats['attempted_three_pointers'] = attempted_three_pointers
nba_stats['made_free_throws'] = made_free_throws
nba_stats['attempted_free_throws'] = attempted_free_throws
nba_stats['offensive_rebounds'] = offensive_rebounds
nba_stats['defensive_rebounds'] = defensive_rebounds
nba_stats['assists'] = assists
nba_stats['steals'] = steals
nba_stats['blocks'] = blocks
nba_stats['turnovers'] = turnovers
nba_stats['personal_fouls'] = personal_fouls
```
